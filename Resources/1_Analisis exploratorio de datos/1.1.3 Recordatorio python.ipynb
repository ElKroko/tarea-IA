**Ciclos**

Existen 2 formas de hacer ciclos: while y for.

Sintaxis del while:



```
while condición:
    instrucciones mientras la condición sea verdadera
```

Sintaxis del for:

```
for variable in secuencia:
    instrucciones
```

el "for" itera sobre la variable "secuencia". Esta variable puede ser un string, alguna estructura de datos (listas, diccionarios, arreglos, etc) o algun archivo.

````python
# Ejemplo de while
cant = int(input("Ingrese la cantidad de alumnos: "))

cont = 0 #contador de alumnos/notas ingresadas
suma = 0 #notas acumuladas

while cont < cant:
    nota = int(input("Ingrese nota: "))
    suma = suma + nota
    cont = cont + 1

print("El promedio es:", int(round(suma/cont)))
````

````python
# Ejemplo de for
s = "hola mundo!"
for caracter in s:
    print("caracter = " + caracter) # podria ser tambien: print('caracter =', caracter)
````

**Funciones**

Se debe usar la palabra reservada "def" para crear la funcion y "return" para retornar el valor. 

Luego de creada la función, puede ser llamada de un código principal o main.

````python
#Ejemplo de creación y uso de una función.
def promedio(p1, p2, p3):
    suma = p1 + p2 + p3
    prom = suma / 3
    return prom

n1 = float(input('Ingrese la primera nota: '))
n2 = float(input('Ingrese la segunda nota: '))
n3 = float(input('Ingrese la tercera nota: '))

prom = promedio(n1, n2, n3) # llamado de la función
print('El promedio es:', prom)
````

El módulo que provee las estructuras de datos y las funciones para trabajar con arreglos se llama NumPy, y no viene incluído con Python, por lo que hay que instalarlo por separado.

El tipo de datos de los arreglos se llama array. Para crear un arreglo nuevo, se puede usar la función "array" pasándole como parámetro la lista de valores que deseamos agregar al arreglo:

Hay muchas formas de arreglos que aparecen a menudo en la práctica, por lo que existen funciones especiales para crearlos:



*   zeros(n) crea un arreglo de n ceros;
*   ones(n) crea un arreglo de n unos;

# Parte 2: ejercicio

Vamos a calcular las distintas frecuencias de un conjunto de datos.


````python
# Vamos a importar un modulo de la libreria de numby llamada random.
from numpy import random

# Vamos a crear datos aleatorios, donde usted decidiráá el largo de estos datos
n = int(input('Ingrese la cantidad de datos: '))

arreglo = np.zeros(n) # vamos a crear un arreglo del tamaño pedido.

i = 0 
while i < len(arreglo):
  numero = random.randint(1,10) # un valor entero aleatorio entre 1 y 10
  arreglo[i] = numero # asignamos el valor en la posicióón i-esima del arreglo
  i = i+1

arreglo
````

````python
# calcularemos la frecuencia absoluta de cada número 
d_fa = {}  # creamos un diccionario para contar cuantos números hay de cada uno. La llave será el número y el valor será las veces que aparece ese número en el arreglo.
L = []  # creamos una lista que nos servira para mostrar los datos en orden gracias a la funcion .sort()

for numero in arreglo: # recorremos el arreglo
  if numero not in d_fa: # si el número no existe como llave del diccionario
    d_fa[numero] = 0     # creamos la llave correspondiente al número. le damos valor 0
    L.append(numero)  # agregamos el numero a la lista. Aca nunca habrán 2 números repetidos.
  d_fa[numero] = d_fa[numero] + 1 # aumentamos en 1 el contador asociado al valor de la llave

d_fa
````

````python
# calculamos la frecuencia relativa de cada número
d_fr = {}
for num, frecuencia in d_fa.items():
  d_fr[num] = frecuencia/n

d_fr
````

````python
# calcularemos las frecuencias acumuladas. Para eso ordenaremos los números de menor a mayor.
L.sort()  # nos aseguramos de tener los números en orden ascendente

d_faa = {}
d_fra = {}

c = 0
for num in L:
  valor = d_fa[num] + c 
  d_faa[num] = valor
  d_fra[num] = valor/n 
  c = valor

print('Diccionario de frecuencias absolutas acumuladas:')
print(d_faa)
print('Diccionario de frecuencias relativas acumuladas:')
print(d_fra)
````

````python
# mostramos todos los datos

print('Valor \t Frecuencia absoluta \t Frecuencia Relativa \t Frecuencia absoluta acumulada \t Frecuencia relativa acumulada')
for num in L:
  print(num,'\t\t', d_fa[num], '\t\t\t', d_fr[num], '\t\t\t   ', d_faa[num], '\t\t\t\t   ', d_fra[num])
````
